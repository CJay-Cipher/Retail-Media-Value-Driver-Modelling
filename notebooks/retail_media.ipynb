```python
# generating dummy brands
brands = {
    "Apple": ["iPhone", "iPad", "Mac", "Apple Watch", "AirPods"],
    "Samsung": ["Galaxy S23", "Galaxy Z Fold 4", "Galaxy Tab S8", "Galaxy Watch 5", "Galaxy Buds Pro"],
    "Nike": ["Air Jordan 1", "Air Force 1", "Nike Zoom Pegasus 39", "Nike Dri-FIT", "Nike Pro"],
    "Adidas": ["Stan Smith", "Superstar", "Ultraboost 23", "Adidas Originals", "Adidas Performance"],
    "Sony": ["PlayStation 5", "Sony WH-1000XM5", "Sony SRS-XG300", "Sony Alpha 7 IV", "Sony ZV-1"],
    "Microsoft": ["Xbox Series X", "Surface Laptop Studio", "Microsoft Office", "Microsoft Azure", "Xbox Game Pass"],
    "Google": ["Pixel 7", "Pixel 6a", "Chromebook", "Google Nest Hub", "Google Workspace"],
    "Coca-Cola": ["Coca-Cola Classic", "Diet Coke", "Coca-Cola Zero Sugar", "Sprite", "Fanta"],
    "Pepsi": ["Pepsi-Cola", "Diet Pepsi", "Pepsi Max", "Mountain Dew", "Gatorade"],
    "Toyota": ["Camry", "Corolla", "RAV4", "Tacoma", "Tundra"],
    "Honda": ["Accord", "Civic", "CR-V", "Pilot", "Ridgeline"],
    "Ford": ["F-150", "Maverick", "Escape", "Bronco", "Mustang"],
    "BMW": ["3 Series", "5 Series", "X3", "X5", "M3"],
    "Mercedes-Benz": ["C-Class", "E-Class", "S-Class", "GLE", "GLS"],
    "Lego": ["Technic", "City", "Star Wars", "Creator Expert", "Ninjago"],
    "Puma": ["Clyde", "Suede", "Future Rider", "Mayze", "Cali Star"],
    "Under Armour": ["Curry Flow 10", "HOVR Phantom 3", "Charged Rogue 3", "Meridian Slide 3", "ColdGear Reactor"],
    "Gucci": ["Ace sneakers", "Dionysus bag", "GG Marmont belt", "Horsebit loafers", "Jackie 1961 bag"],
    "Louis Vuitton": ["Neverfull bag", "Speedy bag", "Alma bag", "Pochette Metis bag", "Sitelle bag"],
    "Rolex": ["Submariner", "Daytona", "GMT-Master II", "Explorer I", "Datejust"],
    "Canon": ["EOS R5", "EOS R6", "EOS 5D Mark IV", "EOS 90D", "PowerShot G7 X Mark III"],
    "Nikon": ["Z 9", "Z 7 II", "Z 6 II", "D850", "D780"],
    "GoPro": ["Hero11 Black", "Hero11 Silver", "Hero11 Mini", "Max Lens Mod", "Media Mod"],
    "Dell": ["XPS 13", "Latitude 7420", "Alienware X17", "Inspiron 15 3000", "G3 15"],
    "HP": ["Spectre x360 14", "Envy 13", "Pavilion 15", "Omen 15", "ZBook Firefly 14c"],
    "Lenovo": ["ThinkPad X1 Carbon", "Yoga 9i", "IdeaPad 5 Pro", "Legion 5 Pro", "ThinkPad P15v Gen 4"],
    "LG": ["C1 OLED TV", "G2 OLED TV", "NanoCell 85 Series", "QNED90 Series", "Soundbar S95QR"],
    "Panasonic": ["GX850 OLED TV", "HZ1500 LED TV", "LUMIX GH6", "LUMIX S5", "Technics SL-1200GR Turntable"],
    "Philips": ["Hue Smart Bulb", "Sonicare DiamondClean Smart 9500", "AirFryer XXL", "OneBlade Pro", "Lumea IPL Hair Removal Device"],
    "Nestle": ["KitKat", "Smarties", "Nespresso", "Aero", "Purina Cat Chow"],
    "Colgate": ["Optic White Toothpaste", "Total Toothbrush", "Sensitive Gum Care", "Max White Enamel Gel", "Luminous White Advanced Teeth Whitening System"],
    "Gillette": ["ProGlide Power Razor", "Fusion5 Razor", "Mach3 Razor", "SkinGuard Razor", "Venus Razor"],
    "Oreo": ["Original Cookies", "Golden Cookies", "Double Stuf Cookies", "Thins Cookies", "Mega Stuf Cookies"],
    "Hershey's": ["Chocolate Bar", "Kisses", "Almond Joy", "Mounds", "Twizzlers"],
    "Crayola": ["Classic Crayons", "Markers", "Colored Pencils", "Paints", "Construction Paper"],
    "PlayStation": ["PlayStation 5 Console", "DualSense Wireless Controller", "PlayStation Plus Subscription", "PlayStation Store Gift Card", "PlayStation Exclusive Games"],
    "Xbox": ["Xbox Series X Console", "Xbox Wireless Controller", "Xbox Game Pass Ultimate Subscription", "Xbox Store Gift Card", "Xbox Exclusive Games"],
    "Nintendo": ["Nintendo Switch Console", "Joy-Con Controllers", "Nintendo Switch Online Subscription", "Nintendo eShop Gift Card", "Nintendo Exclusive Games"],
    "Wrangler": ["Jeans", "Jackets", "Shirts", "Shorts", "Boots"],
    "Vans": ["Classic Slip-On Sneakers", "Old Skool Sneakers", "Authentic Sneakers", "Sk8-Hi Sneakers", "Era Sneakers"],
    "Converse": ["Chuck Taylor All-Stars Sneakers", "Jack Purcell Sneakers", "One Star Sneakers", "Pro Leather Sneakers", "Chuck 70 Sneakers"],
    "Ray-Ban": ["Aviator Sunglasses", "Wayfarer Sunglasses", "Clubmaster Sunglasses", "Round Metal Sunglasses", "Hexagonal Sunglasses"],
    "Oakley": ["Flak Jacket Sunglasses", "Radar EV Sunglasses", "Juliet Sunglasses", "Frogskins Sunglasses", "Gascan Sunglasses"],
    "AmazonBasics": ["AA Rechargeable Batteries", "USB-C Chargers", "HDMI Cables", "Basic T-Shirts", "Laptop Cases"],
    "KitchenAid": ["Stand Mixers", "Hand Mixers, Blenders", "Food Processors", "Coffee Grinders", "Toasters"],
    "Dyson": ["Vacuum Cleaners", "Air Purifiers", "Hair Dryers", "Styling Tools", "Lighting"],
    "Fitbit": ["Charge 5 Fitness Tracker", "Inspire 3 Fitness Tracker", "Versa 4 Smartwatch", "Sense 2 Smartwatch", "Luxe Fitness Tracker"],
    "Sonos": ["One Speaker", "Beam Soundbar", "Arc Soundbar", "Subwoofer", "Roam Speaker"],
    "Bose": ["QuietComfort 45 Headphones", "SoundLink Mini Bluetooth Speaker", "SoundLink Color II Bluetooth Speaker", "Frames Sunglasses with Audio", "SoundSport Bluetooth Headphones"]
}

# Update tag_name values based on brand_name
brand_tag_mapping = {
    "Apple": np.random.choice(brands["Apple"]),
    "Samsung": np.random.choice(brands["Samsung"]),
    "Nike": np.random.choice(brands["Nike"]),
    "Adidas": np.random.choice(brands["Adidas"]),
    "Sony": np.random.choice(brands["Sony"]),
    "Microsoft": np.random.choice(brands["Microsoft"]),
    "Google": np.random.choice(brands["Google"]),
    "Coca-Cola": np.random.choice(brands["Coca-Cola"]),
    "Pepsi": np.random.choice(brands["Pepsi"]),
    "Toyota": np.random.choice(brands["Toyota"]),
    "Honda": np.random.choice(brands["Honda"]),
    "Ford": np.random.choice(brands["Ford"]),
    "BMW": np.random.choice(brands["BMW"]),
    "Mercedes-Benz": np.random.choice(brands["Mercedes-Benz"]),
    "Lego": np.random.choice(brands["Lego"]),
    "Puma": np.random.choice(brands["Puma"]),
    "Under Armour": np.random.choice(brands["Under Armour"]),
    "Gucci": np.random.choice(brands["Gucci"]),
    "Louis Vuitton": np.random.choice(brands["Louis Vuitton"]),
    "Rolex": np.random.choice(brands["Rolex"]),
    "Canon": np.random.choice(brands["Canon"]),
    "Nikon": np.random.choice(brands["Nikon"]),
    "GoPro": np.random.choice(brands["GoPro"]),
    "Dell": np.random.choice(brands["Dell"]),
    "HP": np.random.choice(brands["HP"]),
    "Lenovo": np.random.choice(brands["Lenovo"]),
    "LG": np.random.choice(brands["LG"]),
    "Panasonic": np.random.choice(brands["Panasonic"]),
    "Philips": np.random.choice(brands["Philips"]),
    "Nestle": np.random.choice(brands["Nestle"]),
    "Colgate": np.random.choice(brands["Colgate"]),
    "Gillette": np.random.choice(brands["Gillette"]),
    "Oreo": np.random.choice(brands["Oreo"]),
    "Hershey's": np.random.choice(brands["Hershey's"]),
    "Crayola": np.random.choice(brands["Crayola"]),
    "PlayStation": np.random.choice(brands["PlayStation"]),
    "Xbox": np.random.choice(brands["Xbox"]),
    "Nintendo": np.random.choice(brands["Nintendo"]),
    "Wrangler": np.random.choice(brands["Wrangler"]),
    "Vans": np.random.choice(brands["Vans"]),
    "Converse": np.random.choice(brands["Converse"]),
    "Ray-Ban": np.random.choice(brands["Ray-Ban"]),
    "Oakley": np.random.choice(brands["Oakley"]),
    "AmazonBasics": np.random.choice(brands["AmazonBasics"]),
    "KitchenAid": np.random.choice(brands["KitchenAid"]),
    "Dyson": np.random.choice(brands["Dyson"]),
    "Fitbit": np.random.choice(brands["Fitbit"]),
    "Sonos": np.random.choice(brands["Sonos"]),
    "Bose": np.random.choice(brands["Bose"])
}

all_tags = []
for tag in brands.values():
    all_tags += tag
# Define the feature options
device_type = ['Desktop', 'Mobile', 'Tablet']

advertiser_type = ['Sellers', 'Vendors', 'Non-endemic']

advertiser_size = ['Strategic Accounts', 'Emerging Accounts', 'Longtail']

ad_categories = [
    'Books', 'Films, TV, Music & Games', 'Electronics & Computers', 'Home, Garden, Pets & DIY',
    'Toys, Children & Baby', 'Clothes, Shoes, Jewellery & Accessories', 'Sports & Outdoors',
    'Food & Grocery', 'Health & Beauty', 'Car & Motorbike'
]
retailer_names = [
    'Amazon', 'AliExpress' #, 'eBay', 'Walmart', 'Alibaba', 'Target', 'Best Buy', 'Etsy', 'Google', 'Microsoft'
]

ad_brands = [
    'Apple', 'Samsung', 'Nike', 'Adidas', 'Sony', 'Microsoft', 'Google', 'Coca-Cola', 'Pepsi', 'Toyota',
    'Honda', 'Ford', 'BMW', 'Mercedes-Benz', 'Lego', 'Puma', 'Under Armour', 'Gucci', 'Louis Vuitton',
    'Rolex', 'Canon', 'Nikon', 'GoPro', 'Dell', 'HP', 'Lenovo', 'LG', 'Panasonic', 'Philips', 'Nestle', 'Colgate',
    'Gillette', 'Oreo', 'Hershey\'s', 'Crayola', 'PlayStation', 'Xbox', 'Nintendo', 'Wrangler', 'Vans', 'Converse',
    'Ray-Ban', 'Oakley', 'AmazonBasics', 'KitchenAid', 'Dyson', 'Fitbit', 'Sonos', 'Bose'
]


# Generate random data
np.random.seed(42)

# Categories and corresponding search keywords
categories = {
    'Books': ['fiction book', 'non-fiction book', 'mystery novel', 'romance novel'],
    'Films, TV, Music & Games': ['action movie', 'drama TV series', 'comedy film', 'adventure game'],
    'Electronics & Computers': ['smartphone', 'laptop', 'headphones', 'camera', 'tablet'],
    'Home, Garden, Pets & DIY': ['sofa', 'garden tools', 'pet food', 'paint'],
    'Toys, Children & Baby': ['toy robot', 'baby stroller', 'educational toy', 'board game'],
    'Clothes, Shoes, Jewellery & Accessories': ["men's shoes", "women's dress", 'necklace', 'hat'],
    'Sports & Outdoors': ['running shoes', 'basketball', 'outdoor tent', 'yoga mat'],
    'Food & Grocery': ['chocolate', 'coffee', 'cereal', 'olive oil'],
    'Health & Beauty': ['face moisturizer', 'hair dryer', 'perfume', 'vitamin supplement'],
    'Car & Motorbike': ['car tires', 'motorcycle helmet', 'car wax', 'bike lock']
}

# --------------------------------------------------------------------------------------------------------------------------------
"""
To modify the dataset values, just update the supply_demand dictionary below
Each key contains the following values respectively
[Ad Coverage, Cost Per Click, Viewable Impression, Ad Density, Total Advertisers, Activated Advertiser %]
"""
supply_demand = {
    "jan": [0.85, 0.67, 168_300, 0.020, 125, 0.20],
    "feb": [0.84, 0.58, 128_898, 0.020, 125, 0.15],
    "mar": [0.86, 0.63, 158_882, 0.020, 125, 0.17],
    "apr": [0.88, 0.65, 186_595, 0.020, 125, 0.19],
    "may": [0.84, 0.66, 193_687, 0.020, 125, 0.21],
    "jun": [0.86, 0.64, 221_783, 0.022, 125, 0.23],
    "jul": [0.88, 0.60, 191_505, 0.022, 125, 0.18],
    "aug": [0.90, 0.58, 177_876, 0.024, 125, 0.16],
    "sep": [0.89, 0.64, 259_524, 0.023, 125, 0.24],
    "oct": [0.87, 0.68, 325_728, 0.025, 125, 0.32],
    "nov": [0.90, 0.72, 481_140, 0.026, 125, 0.44],
    "dec": [0.91, 0.78, 687_960, 0.027, 125, 0.60]
}
months = ["{:02d}".format(x) for x in range(1, 13)]

for i, month in enumerate(supply_demand.values()):
    num_true = month[2]  # Viewable Impressions
    ad_covered = month[0]  # Ad Covered
    num_rows = int(num_true / ad_covered)

    ctr = month[3]
    cpc = month[1]
    total_advertisers = month[4]  # -------> Total Advertisers
    activated_advertisers = month[5]

    month_days = 31
    if i == 1:
        month_days = 28
    elif i in [3, 5, 8, 10]:
        month_days = 30

    # Impression Table
    impression_data = {
        'datetime': pd.date_range(start=f'2022-{months[i]}-01', end=f'2022-{months[i]}-{month_days}', periods=num_rows).floor('S'),
        'item_category': np.random.choice(list(categories.keys()), size=num_rows),
        'ad_id': np.random.randint(1, len(all_tags), size=num_rows),
        'user_id': np.random.randint(1, 500, size=num_rows),
        'retailer_id': np.random.randint(1, len(retailer_names), size=num_rows),
        'was_viewed': [True] * num_true + [False] * (num_rows - num_true),  # Generate exactly num_true True values
        'was_clicked': np.random.choice([True, False], size=num_rows),
        'page_number': np.random.randint(1, 10, size=num_rows),
        'device_id': np.random.randint(1, len(device_type)+1, size=num_rows)
    }

    impression_df = pd.DataFrame(impression_data)

    # Update was_clicked to False if was_viewed is False
    impression_df.loc[impression_df['was_viewed'] == False, 'was_clicked'] = False

    # Calculate the count of True values in was_viewed
    true_count = impression_df['was_viewed'].sum()

    # Calculate the desired count of True values in was_clicked based on 2%
    count_true_clicked = int(true_count * ctr)  # -------> CTR

    # Get the indices of True values in was_viewed
    true_indices = impression_df[impression_df['was_viewed'] == True].index

    # Randomly select count_true_clicked indices and update the corresponding values in was_clicked to True
    indices_to_update = np.random.choice(true_indices, count_true_clicked, replace=False)
    impression_df['was_clicked'] = False
    impression_df.loc[indices_to_update, 'was_clicked'] = True

    # Generate search keywords based on the selected category
    impression_df['search_keyword'] = \
        impression_df['item_category'].apply(lambda x: random.choice(categories[x]))


    # Ad table ------------------------------------------------------------------------------------------------
    target_average = cpc  # -------> CPC
    while True:
        array = np.random.uniform(0, target_average * 2, size=len(all_tags))
        if np.abs(np.mean(array) - target_average) < 0.0001:
            break

    # Corrected cost_per_click values
    cost_per_click = np.round(array, 2) 

    numbers = list(range(1, total_advertisers + 1))
    repetitions = len(all_tags) // len(numbers)
    remainder = len(all_tags) % len(numbers)
    advertiser_list = numbers * repetitions + numbers[:remainder]

    ad_data = {
        'ad_id': range(1, len(all_tags)+1),
        'tag_name': all_tags,
        'advertiser_id': advertiser_list,
        'ad_brand': np.random.choice(list(brand_tag_mapping.keys()), size=len(all_tags), replace=True),
        'cost_per_click': cost_per_click
    }

    ad_df = pd.DataFrame(ad_data)

    brand_mapping = {bran: i+1 for i, bran in enumerate(ad_brands)}
    ad_df['brand_id'] = ad_df['ad_brand'].map(brand_mapping)
    ad_df = ad_df.drop('ad_brand', axis=1)


    # Advertiser table -----------------------------------------------------------------------------------------------
    
    start_date = pd.to_datetime('2020-01-01')
    end_date = pd.to_datetime('2021-12-31')
    date_range = pd.date_range(start=start_date, end=end_date, periods=total_advertisers).floor('S')
    random_dates = random.sample(date_range.tolist(), total_advertisers)

    advertiser_data = {
        'advertiser_id': np.arange(1, total_advertisers + 1),
        'advertiser_name': ['Advertiser ' + str(i) for i in range(1, total_advertisers + 1)],
        'advertiser_country': np.random.choice(['USA', 'UK', 'Canada'], size=total_advertisers),
        'advertiser_type': np.random.choice(advertiser_type, size=total_advertisers),
        'advertiser_size': np.random.choice(advertiser_size, size=total_advertisers),
        'is_activated': False,
        'date_registered': random_dates
    }

    advertiser_df = pd.DataFrame(advertiser_data)

    # Modify is_activated column
    desired_true_count = int(total_advertisers * activated_advertisers)  # -------> Activated Advertisers
    random_indices = np.random.choice(total_advertisers, size=desired_true_count, replace=False)
    advertiser_df.loc[random_indices, 'is_activated'] = True


    # Merge the dataframes
    merged_df = pd.merge(impression_df, ad_df, on='ad_id', how='left')
    # merged_df = pd.merge(merged_df, user_df, on='user_id', how='left')
    # merged_df = pd.merge(merged_df, device_df, on='device_id', how='left')
    merged_df = pd.merge(merged_df, advertiser_df, on='advertiser_id', how='left')
    # merged_df = pd.merge(merged_df, retailer_df, on='retailer_id', how='left')
    # merged_df = pd.merge(merged_df, brand_df, on='brand_id', how='left')

    # Save the merged dataframe to a CSV file
    merged_df.to_csv(f'dummy_data/merged_data{i+1}.csv', index=False)

    
# User table ------------------------------------------------------------------------------------------------
user_data = {
    'user_id': np.arange(1, 501),  # Unique user_id as primary key
    'username': ['User ' + str(i) for i in range(1, 501)],
    'is_registered': np.random.choice([True, False], size=500),
    'ip_address': ['192.168.0.' + str(i) for i in range(1, 501)],
    'country': np.random.choice(['USA', 'UK', 'Canada'], size=500)
}
user_df = pd.DataFrame(user_data)

# Device table ------------------------------------------------------------------------------------------------
device_data = {
    'device_id': [1, 2, 3],
    'device_type': device_type
}
device_df = pd.DataFrame(device_data)

# # Retailer table ------------------------------------------------------------------------------------------------
# retailer_data = {
#     'retailer_id': list(range(1, len(retailer_names)+1)),  # Unique retailer_id as primary key
#     'retailer_name': retailer_names
# }
# retailer_df = pd.DataFrame(retailer_data)

# Brand table ------------------------------------------------------------------------------------------------
brand_data = {
    'brand_id': list(range(1, len(ad_brands)+1)),  # Unique brand_id as primary key
    'brand_name': ad_brands
}
brand_df = pd.DataFrame(brand_data)

# impression_df.to_csv(f'dummy_data/impression_data{i}.csv', index=False)
# ad_df.to_csv(f'dummy_data/ad_data{i}.csv', index=False)
user_df.to_csv(f'dummy_data/user_data.csv', index=False)
device_df.to_csv(f'dummy_data/device_data.csv', index=False)
# advertiser_df.to_csv(f'dummy_data/advertiser_data{i}.csv', index=False)
# retailer_df.to_csv(f'dummy_data/retailer_data{i}.csv', index=False)
brand_df.to_csv(f'dummy_data/brand_data.csv', index=False)



# Create an empty DataFrame to store the merged data
merged_data = pd.DataFrame()
# total = 0

# Loop through the CSV files and vertically concatenate them
for i in range(1, 13):
    file_path = f"dummy_data/merged_data{i}.csv"
    df = pd.read_csv(file_path)
    # total += df.shape[0]
    df.reset_index(drop=True, inplace=True)  # Reset the index for each dataframe
    merged_data = pd.concat([merged_data, df], axis=0)
    del df  # Delete the DataFrame after reading it
    os.remove(file_path)  # Delete the file

# Continue with further operations on merged_data

# Reset the index of the merged data
merged_data.reset_index(drop=True, inplace=True)

print("Shape -", merged_data.shape, "\n")
# print(total)

merged_data.to_csv(f'dummy_data/all_data.csv', index=False)
print("*** All dataset Successfully Saved ***")

merged_data.head()
```
