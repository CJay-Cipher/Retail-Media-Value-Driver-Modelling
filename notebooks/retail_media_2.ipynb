#### Generate data for each month then merge them together

```python

"""
To modify the dataset values, just update the monthly_values dictionary below
Each key contains the following values respectively
[Ad Coverage, Cost Per Click, Viewable Impression, CTR, Total Advertisers, Activated Advertiser %]
"""
monthly_values = {
    "jan": [0.85, 0.67, 168_300, 0.0201, 129, 0.20],
    "feb": [0.84, 0.58, 128_898, 0.0211, 123, 0.15],
    "mar": [0.86, 0.63, 158_882, 0.0221, 132, 0.17],
    "apr": [0.88, 0.65, 186_595, 0.0232, 125, 0.19],
    "may": [0.84, 0.66, 193_687, 0.0205, 129, 0.21],
    "jun": [0.86, 0.64, 221_783, 0.0223, 131, 0.23],
    "jul": [0.88, 0.60, 191_505, 0.0225, 127, 0.18],
    "aug": [0.90, 0.58, 177_876, 0.0242, 133, 0.16],
    "sep": [0.89, 0.64, 259_524, 0.0231, 126, 0.24],
    "oct": [0.87, 0.68, 325_728, 0.0253, 134, 0.32],
    "nov": [0.90, 0.72, 481_140, 0.0262, 137, 0.44],
    "dec": [0.91, 0.78, 687_960, 0.0274, 139, 0.60]
}
month_nums = [f"{x:02d}" for x in range(1, 13)]  # ['01', '02', '03', '04', '05', '06', '07', '08', '09', '10', '11', '12']

for i, month in enumerate(monthly_values.values()): 
    ad_covered = month[0] 
    avg_cost_per_click = month[1]
    view_impressions = month[2] 
    total_rows = int(view_impressions / ad_covered)
    click_through_rate = month[3]
    total_advertisers = month[4] 
    activated_advertisers_prtg = month[5]

    # Determine the number of days in the month based on the month index
    month_days = 31
    if i == 1:
        month_days = 28
    elif i in [3, 5, 8, 10]:
        month_days = 30

    # Generate impression data for each month
    impression_data = {
        'datetime': pd.date_range(start=f'2022-{month_nums[i]}-01', end=f'2022-{month_nums[i]}-{month_days}', periods=total_rows).floor('S'),
        'item_category': np.random.choice(list(categories.keys()), size=total_rows),
        'ad_id': np.random.randint(1, len(all_ad_tags), size=total_rows),
        'user_id': np.random.randint(1, 500, size=total_rows),
        'retailer_id': np.random.randint(1, len(retailer_names), size=total_rows),
        'was_viewed': [True] * view_impressions + [False] * (total_rows - view_impressions), 
        'was_clicked': np.random.choice([True, False], size=total_rows),
        'page_number': np.random.randint(1, 10, size=total_rows),
        'device_id': np.random.randint(1, len(device_type)+1, size=total_rows)
    }
    impression_df = pd.DataFrame(impression_data)

    # Update was_clicked to False if was_viewed is False
    impression_df.loc[impression_df['was_viewed'] == False, 'was_clicked'] = False

    # Calculate the count of True values in was_viewed
    true_count = impression_df['was_viewed'].sum()

    # Calculate the desired count of True values in was_clicked based on ctr %
    count_true_clicked = int(true_count * click_through_rate)  # -------> CTR

    # Get the indices of True values in was_viewed
    true_indices = impression_df[impression_df['was_viewed'] == True].index

    # Randomly select count_true_clicked indices and update the corresponding values in was_clicked to True
    indices_to_update = np.random.choice(true_indices, count_true_clicked, replace=False)
    impression_df['was_clicked'] = False
    impression_df.loc[indices_to_update, 'was_clicked'] = True

    # Generate search keywords based on the selected category
    impression_df['search_keyword'] = \
        impression_df['item_category'].apply(lambda x: random.choice(categories[x]))


    # Ad table ------------------------------------------------------------------------------------------------
    # Generate an array of random values for cost per click
    while True:
        array = np.random.uniform(0, avg_cost_per_click * 2, size=len(all_ad_tags))
        if np.abs(np.mean(array) - avg_cost_per_click) < 0.0001:
            break
    cost_per_click = np.round(array, 2)  # Round the values to two decimal places

    # generate list of advertizers to attain total monthly advertizers
    numbers = list(range(1, total_advertisers + 1))
    repetitions = len(all_ad_tags) // len(numbers)
    remainder = len(all_ad_tags) % len(numbers)
    advertiser_list = numbers * repetitions + numbers[:remainder]
    
    # Generate ad data
    ad_data = {
        'ad_id': range(1, len(all_ad_tags)+1),
        'tag_name': all_ad_tags,
        'advertiser_id': advertiser_list,
        'ad_brand': np.random.choice(list(brands.keys()), size=len(all_ad_tags), replace=True),
        'cost_per_click': cost_per_click
    }

    ad_df = pd.DataFrame(ad_data)

    # add brand ID for each ad brand
    brand_mapping = {bran: i+1 for i, bran in enumerate(ad_brands)}
    ad_df['brand_id'] = ad_df['ad_brand'].map(brand_mapping)
    ad_df = ad_df.drop('ad_brand', axis=1)
    

    # Advertiser table -----------------------------------------------------------------------------------------------
    # Generate advertiser data
    start_date = pd.to_datetime('2020-01-01')
    end_date = pd.to_datetime('2021-12-31')
    date_range = pd.date_range(start=start_date, end=end_date, periods=total_advertisers).floor('S')
    random_dates = random.sample(date_range.tolist(), total_advertisers)

    advertiser_data = {
        'advertiser_id': np.arange(1, total_advertisers + 1),
        'advertiser_name': ['Advertiser ' + str(i) for i in range(1, total_advertisers + 1)],
        'advertiser_country': np.random.choice(['USA', 'UK', 'Canada'], size=total_advertisers),
        'advertiser_type': np.random.choice(advertiser_type, size=total_advertisers),
        'advertiser_size': np.random.choice(advertiser_size, size=total_advertisers),
        'is_activated': False,
        'date_registered': random_dates
    }

    advertiser_df = pd.DataFrame(advertiser_data)

    # Modify is_activated advertizes to meet monthly expectation
    desired_true_count = int(total_advertisers * activated_advertisers_prtg) 
    random_indices = np.random.choice(total_advertisers, size=desired_true_count, replace=False)
    advertiser_df.loc[random_indices, 'is_activated'] = True

    # Merge the impression, ad and advertizer tables
    merged_df = pd.merge(impression_df, ad_df, on='ad_id', how='left')
    merged_df = pd.merge(merged_df, advertiser_df, on='advertiser_id', how='left')

    # Save the merged dataframe to a CSV file
    merged_df.to_csv(f'dummy_data/merged_data{i+1}.csv', index=False)

```

```python
merged_data = pd.DataFrame()  # Create an empty DataFrame named 'merged_data'

# Loop through the CSV files and vertically concatenate them
for i in range(1, 13):
    file_path = f"dummy_data/merged_data{i}.csv"  # Generate the file path for merged data file with index i
    df = pd.read_csv(file_path)  # Read the CSV file into a DataFrame
    df.reset_index(drop=True, inplace=True)  # Reset the index of the DataFrame
    merged_data = pd.concat([merged_data, df], axis=0)  # Concatenate the current DataFrame with the merged data
    del df  # Delete the temporary DataFrame to free up memory
    os.remove(file_path)  # Remove the CSV file from the file system

# Reset the index of the merged data
merged_data.reset_index(drop=True, inplace=True)

print("Shape -", merged_data.shape, "\n")

merged_data.to_csv(f'dummy_data/all_data.csv', index=False)
print("*** All dataset Successfully Saved ***")

merged_data.head()
```
